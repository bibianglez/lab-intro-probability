**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
import scipy.stats as stats
```

```python
Sold_tickets = 460
available_seats = 450
probability = 1-0.03

success_probability = stats.binom.cdf(available_seats, Sold_tickets, probability)

print("--- Results ---")
print(f"The probability for success is: {success_probability:.4f}")
print(f"in percentaje: {success_probability * 100:.2f}%")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
p = 0.3

final_probability = 1 - stats.geom.cdf(2, p)

print(f"The probability is: {final_probability:.4f}")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
visit_mean = 500
capacity = 550

overwhelm_posibility = 1 - stats.poisson.cdf(capacity , visit_mean)

print("--- Results ---")
print(f"The probability of the server being overwhelmed is: {overwhelm_posibility:.4f}")
print(f"In percentage: {overwhelm_posibility * 100:.2f}%")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
visit_mean_24 = 500 * 24
capacity_24 = 550 * 24

overwhelm_posibility_24 = 1 - stats.poisson.cdf(capacity_24 , visit_mean_24 )

print("--- Results ---")
print(f"The probability of the server being overwhelmed is: {overwhelm_posibility_24:.4f}")
print(f"In percentage: {overwhelm_posibility_24 * 100:.2f}%")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
avg_time = 10 

next_time = 5

arrival_probability = stats.expon.cdf(next_time, scale=avg_time)

print("--- Results ---")
print(f"The probability that the next customer arrives within 5 minutes is: {arrival_probability:.4f}")
print(f"In percentage: {arrival_probability * 100:.2f}%")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
avg_time = 10 

time_break = 15

break_probability = 1 - stats.expon.cdf(time_break, scale=avg_time)

print("--- Results ---")
print(f"The probability of an employee taking a break is: {break_probability:.4f}")
print(f"In percentage: {break_probability * 100:.2f}%")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
mean_weight = 150
std_dev = 10

Inf_limit = 140
high_limit = 160

range_probability = stats.norm.cdf(high_limit, loc=mean_weight, scale=std_dev) - \
                     stats.norm.cdf(Inf_limit, loc=mean_weight, scale=std_dev)


print("--- Results ---")
print(f"The probability that the bird weighs between 140 and 160 grams is: {range_probability:.4f}")
print(f"In percentage: {range_probability * 100:.2f}%")
```

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
avg_time = 50 

limit_time = 30

probabilidad_llegada = stats.expon.cdf(limit_time, scale=avg_time)

print("--- Results ---")
print(f"The probability is: {probabilidad_llegada:.4f}")
print(f"In percentage: {probabilidad_llegada * 100:.2f}%")
```
